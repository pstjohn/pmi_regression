## Perform OLS regressions

```python
import statsmodels.api as sm
```

### EPAct/V2

```python
epa_data = sm.add_constant(epa[['Density', 'T70', 'YSI HOV VP-1']])
mod = sm.OLS(epa.y, epa_data)
res = mod.fit()
res.summary()
```

### Ratcliff

```python
ratcliff_data = sm.add_constant(ratcliff[['Density', 'T70', 'YSI HOV VP-1']])
mod = sm.OLS(ratcliff.y, ratcliff_data)
res = mod.fit()
res.summary()
```
